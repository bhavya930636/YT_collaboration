```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("channel_and_max_impact_collab.csv")
```

```python
import matplotlib.pyplot as plt

from_categories = df['from_category'].unique()

# Set up grid size
n_categories = len(from_categories)
n_rows, n_cols = 4, 4  # To accommodate up to 16 categories

fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, 16))
axes = axes.flatten()  # Flatten to easily index with a single loop

for i, category in enumerate(from_categories):
    subset = df[df['from_category'] == category]
    to_category_counts = subset['to_category'].value_counts()

    ax = axes[i]
    ax.pie(
        to_category_counts,
        labels=[f"{cat} ({count})" for cat, count in zip(to_category_counts.index, to_category_counts)],
        startangle=140
    )
    ax.set_title(f"'{category}' → Others", fontsize=10)
    ax.axis('equal')

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.suptitle("Collaborations by 'from_category'", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python


# Group by 'from_category'
from_categories = df['from_category'].unique()

for category in from_categories:
    subset = df[df['from_category'] == category]
    
    # Pie chart showing count of collaborations to each 'to_category'
    to_category_counts = subset['to_category'].value_counts()
    
    # Plotting pie chart with actual counts
    plt.figure(figsize=(6, 6))
    plt.pie(to_category_counts, labels=[f"{cat} ({count})" for cat, count in zip(to_category_counts.index, to_category_counts)],
            startangle=140)
    plt.title(f"Collaborations of '{category}' Channels with Other Categories")
    plt.axis('equal')
    plt.tight_layout()
    plt.show()
    
    # Top 10 'to_category' by sum of max_impact
    top10 = (
        subset.groupby('to_category')['max_impact']
        .sum()
        .sort_values(ascending=False)
        .head(10)
    )
    
    print(f"\nTop 10 'to_category' by max_impact for from_category = '{category}':\n")
    print(top10)
    print("\n" + "-"*60 + "\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print(df)
# === Load the data ===
# df = pd.read_csv("your_file.csv")  # Replace with your actual file name

# === Convert popularity values to numeric (they're strings initially) ===
df["from_popularity"] = pd.to_numeric(df["from_popularity"], errors="coerce")
df["to_popularity"] = pd.to_numeric(df["to_popularity"], errors="coerce")
df["max_impact"] = pd.to_numeric(df["max_impact"], errors="coerce")
df["videos"] = pd.to_numeric(df["videos"], errors="coerce")

# Drop rows with missing or invalid values
df = df.dropna(subset=["from_popularity", "to_popularity", "max_impact", "videos"])

# === Derived Features ===
df["popularity_gain"] = df["to_popularity"] - df["from_popularity"]
df["from_is_small"] = df["from_popularity"] < df["from_popularity"].quantile(0.25)
df["is_high_effort"] = df["videos"] > df["videos"].quantile(0.75)

# === Identify MVPs ===
mvp_df = df[
    (df["from_is_small"]) &
    (df["popularity_gain"] > df["popularity_gain"].quantile(0.75)) &
    (df["max_impact"] > df["max_impact"].quantile(0.75))
]

print("🎯 Top Collab MVPs:")
print(mvp_df[["from", "to", "from_popularity", "to_popularity", "popularity_gain", "max_impact"]]
      .sort_values(by="popularity_gain", ascending=False).head(10))

# === Identify Freeriders ===
freeriders_df = df[
    (df["is_high_effort"]) &
    (df["popularity_gain"] < df["popularity_gain"].quantile(0.25)) &
    (df["max_impact"] < df["max_impact"].quantile(0.25))
]

print("\n🪃 Potential Freeriders:")
print(freeriders_df[["from", "to", "videos", "popularity_gain", "max_impact"]]
      .sort_values(by="videos", ascending=False).head(10))
```

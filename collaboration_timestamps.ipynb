```python
import pandas as pd
from datetime import datetime
from dateutil import parser
import re

# Step 1: Read and parse the TXT file
video_ids = []
upload_times = []

with open("all_video_metadata_Bhoomi.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()
fail=0
for i in range(len(lines)):
    if lines[i].startswith("Video ID:"):
        # Extract video ID
        video_id_match = re.search(r"Video ID: (\S+)", lines[i])
        if video_id_match:
            video_id = video_id_match.group(1)
        
        # Extract upload time
        if i + 1 < len(lines) and "Uploaded at:" in lines[i + 1]:
            upload_time_str = lines[i + 1].split("Uploaded at:")[1].strip()
            
            try:
                # Robust parsing (auto handles formats like 'T'/'Z' or space)
                upload_time = parser.parse(upload_time_str)
            except Exception as e:
                fail+=1
                print(f"Failed to parse time for video ID {video_id}: {upload_time_str}")
                upload_time = None

            video_ids.append(video_id)
            upload_times.append(upload_time)
print(fail)
# Step 2: Create DataFrame
video_times = pd.DataFrame({
    'videoID': video_ids,
    'upload_time': upload_times
})

# Convert all times to UTC (if they aren't already)
video_times['upload_time'] = pd.to_datetime(video_times['upload_time'], utc=True)

# Optional: check first few rows
print(video_times)
```

```python
import pandas as pd
import ast

# Example of how video_times DataFrame might look:
# video_times = pd.DataFrame({'videoID': ['6t10zCike1g', '77JsEDZ6EYc', ...], 'upload_time': [timestamp1, timestamp2, ...]})

# Create a lookup dictionary from the video_times DataFrame (make sure video_times is a DataFrame)
video_time_dict = dict(zip(video_times['videoID'], video_times['upload_time']))

# Load collaboration data
collab_df = pd.read_csv("df_collabs.txt", delimiter="\t")  # or use appropriate delimiter
collab_df
```

```python

# Create records only with available timestamps
records = []

for _, row in collab_df.iterrows():
    from_id = row["from"]
    to_id = row["to"]
    
    # Convert the string-like list into an actual list (remove extra quotes)
    video_ids = ast.literal_eval(row["videos"].replace('""', '"'))  # Fix extra quotes
    video_times = []
    # print(video_ids)
    for video_id in video_ids:
        # print(video_id)
        if video_id not in video_time_dict:
            continue  # Skip if timestamp is not available
        
        video_timestamp = video_time_dict[video_id]
        video_times.append(video_timestamp)
    
    # Append the record if video timestamps are available
    if video_times:
        records.append({
            "from": from_id,
            "to": to_id,
            "video_id": video_ids,
            "video_timestamp": video_times
        })

# Convert to DataFrame and inspect
final_df = pd.DataFrame(records)
print(final_df)
```
